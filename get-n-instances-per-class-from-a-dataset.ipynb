```python
import csv

# specify the input and output file paths
input_file = '/kaggle/input/matrix-lsamediapipe/padded_matrix_file (1).csv'
output_file = 'output.csv'

# specify the number of rows to keep per class
rows_per_class = 5

# create a dictionary to store the data for each class
class_data = {}
with open(input_file, 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        class_num = int(row[-1]) # assume the class number is in the last column
        if class_num not in class_data:
            class_data[class_num] = []
        class_data[class_num].append(row)

# write the filtered data to the output file
with open(output_file, 'w', newline='') as f:
    writer = csv.writer(f)
    for class_num, rows in class_data.items():
        # if there are fewer than rows_per_class rows for this class, keep them all
        if len(rows) <= rows_per_class:
            for row in rows:
                writer.writerow(row)
        # otherwise, keep the first rows_per_class rows
        else:
            for i in range(rows_per_class):
                writer.writerow(rows[i])
```
