```python
import os
import subprocess

def trim_video(input_path, start_time, end_time):
    # Create the output folder called "trimmed" if it doesn't exist
    output_folder = "trimmed"
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Create the output file path
    input_file_name = input_path.split("/")[-1].split(".")[0]
    output_file_name = f"{input_file_name}.mp4"
    output_path = f"{output_folder}/{output_file_name}"

    # Run the ffmpeg command to trim the video
    cmd = (
        f"ffmpeg -i {input_path} -ss {start_time} -to {end_time} -async 1 {output_path}"
    )
    subprocess.run(cmd, shell=True, check=True)

    print(f"Trimmed video saved at: {output_path}")

input_path = "/kaggle/input/slsl-2-cropped/slsl-2-cropped/050_005_002.mp4"
start_time = 0  # In seconds
end_time = 3   # In seconds

trim_video(input_path, start_time, end_time)
```
